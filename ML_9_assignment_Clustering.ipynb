```python
## Source Code: https://appliedmachinelearning.blog/2017/03/08/image-compression-using-k-means-clustering/


import numpy as np
import pandas as pd
from sklearn import cluster, datasets
import scipy.misc
```

```python
import matplotlib.pyplot as plt
f= scipy.misc.face(gray=True)
plt.figure(figsize=(10,3.6))
plt.subplot(131)
plt.imshow(f,cmap=plt.cm.gray)
```

## Visualizing the gray scale racoon image

```python
plt.subplot(132)
plt.imshow(f,cmap=plt.cm.gray,vmin=30,vmax=200)
plt.axis('off')
```

```python
plt.imshow(f,cmap=plt.cm.gray)
plt.contour(f,[10,200])
plt.axis('off')
plt.subplots_adjust(wspace=0, hspace=0, top = 0.99,bottom=0.01,left=0.05,right=0.99)
plt.show()
```

```python
f.shape
```

## Applying KMeans Algorithm and clustering for 5 clusters

```python
from skimage import io
from sklearn.cluster import KMeans
```

```python
rows=f.shape[0]
cols=f.shape[1]
```

```python
f.shape
```

```python
k_colors = KMeans(n_clusters=5).fit(f)
y_pred = k_colors.predict(f)
```

```python
k_colors.cluster_centers_.shape
```

```python
newimg=k_colors.cluster_centers_[k_colors.labels_]
```

```python
newimg=np.reshape(newimg,(f.shape))
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,2,1,xticks=[],yticks=[],title='Original Image')
ax.imshow(f)
ax=fig.add_subplot(1,2,2,xticks=[],yticks=[],title='Color Compressed Image using K-means')
ax.imshow(newimg)
plt.show()
```

```python
newimg.shape
```

```python
## compressed gray scale image of racoon
plt.imshow(newimg,cmap=plt.cm.gray)
plt.contour(newimg,[10,200])
plt.axis('off')
plt.subplots_adjust(wspace=0, hspace=0, top = 0.99,bottom=0.01,left=0.05,right=0.99)
plt.show()
```

```python
##Applying Kmeans for 128 clusters and visulaizing the difference in the raw and compressed image


```

```python
k_colors = KMeans(n_clusters=128).fit(f)
y_pred = k_colors.predict(f)

newimg=k_colors.cluster_centers_[k_colors.labels_]
newimg=np.reshape(newimg,(f.shape))

```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,2,1,xticks=[],yticks=[],title='Original Image')
ax.imshow(f)
ax=fig.add_subplot(1,2,2,xticks=[],yticks=[],title='Color Compressed Image using K-means')
ax.imshow(newimg)
plt.show()
```
